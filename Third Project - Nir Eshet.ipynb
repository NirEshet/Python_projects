# Handle null values:

```python
players.isna().sum()
```

```python
players[players["current_price_(k€)"].between(0.02,13)]["current_price_(k€)"].mean().round(2)
```

* current_price_(k€) row - in this code, I took the values between 0.02 (which is the minimum value in the dataset) and 13 (which is 75% in my dataset) from "current_price_(k€)" column. I also did average for all the values and present only 2 numbers after the decimal point. The number (4.01) represent the average current_price for all the values between 0.02 and 13. This is the reason why I decieded to handle the nulls on this row.

```python
players[players["potential_price"].between(0.02,22)]["potential_price"].mean().round(2)
```

* Potential_price row - in this code, I took the values between 0.02 (which is the minimum value in the dataset) and 22 (which is 75% in my dataset) from "potential_price" column. I also did average for all the values and present only 2 numbers after the decimal point. The number (7.69) represent the average current_price for all the values between 0.02 and 22. This is the reason why I decieded to handle the nulls on this row.

* Height row - when we show the statistics of the numeric columns by using "describe" method, I noticed that the standard deviation is very low (0.07) and the value was 1.83, so I used this to fill in the nulls.

* Foot row - just put "No value entered" to show there is no data there.

* Outfitter row - just put "No value entered" to show there is no data there.

* Contract_expires row - I didnt handle the nulls (no need ).

```python
players["height"].fillna(1.83,inplace=True)
players["current_price_(k€)"].fillna(4.01,inplace=True)
players["potential_price"].fillna(7.69,inplace=True)
players["foot"].fillna("No value entered",inplace=True)
players["outfitter"].fillna("No value entered",inplace=True)
```

# Plot Visualizations:

Plot 1: In plots 1-4 I analyzed the dominant foot for the defensive players and in 5 plot I analyzed the dominant foot for all pupulation (all players).

```python
# Plot 1
plt.style.use('seaborn-v0_8-colorblind')
plt.figure(figsize=(10,10))
ax=plt.subplot(3, 2, 1)
sb.histplot(data=players[players['position'] == 'Goalkeeper'], x="foot", hue='foot', kde=True)
plt.title("Dominant Foot - Goalkeeper", fontsize=12)
plt.xlabel("Player Dominant Foot")
plt.ylabel("Count")

# Plot 2
plt.subplot(3, 2, 2,sharey =ax)
sb.histplot(data=players[players['position'] == 'Defender - Centre-Back'], x="foot", hue='foot', kde=True)
plt.title("Dominant Foot - Centre Back", fontsize=12)
plt.xlabel("Player Dominant Foot")
plt.ylabel("Count")

# Plot 3
plt.subplot(3, 2, 3,sharey =ax)
sb.histplot(data=players[players['position'] == 'Defender - Left-Back'], x="foot", hue='foot', kde=True)
plt.title("Dominant Foot - Left Back", fontsize=12)
plt.xlabel("Player Dominant Foot")
plt.ylabel("Count")


# Plot 4
plt.subplot(3, 2, 4,sharey =ax)
sb.histplot(data=players[players['position'] == 'Defender - Right-Back'], x="foot", hue='foot', kde=True)
plt.title("Dominant Foot - Right Back", fontsize=12)
plt.xlabel("Player Dominant Foot")
plt.ylabel("Count")

# Plot 5
plt.subplot(3, 2, 5)
sb.histplot(data=players, x="foot", hue='foot', kde=True)
plt.title("Dominant Foot - All Positions", fontsize=12)
plt.xlabel("Player Dominant Foot")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

Plot 2: In these 2 plots , I analyzed first, how many players belong to each age.

 Player price distrobution by league:

```python
plt.figure(figsize=(8,6))
sb.set_theme(style="darkgrid", palette="Set2")
sb.barplot(data=players, x="league",y="current_price_(k€)",hue="league",estimator="sum",ci=False)
plt.xlabel("League")
plt.ylabel("Current_Price_(k€)")
plt.title("Total Player Value By League")
plt.show()
```
